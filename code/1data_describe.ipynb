## 1. แสดงข้อมูลเชิงสถิติ

### เริ่มจากนำเข้าข้อมูล

```python
import pandas as pd
data_daily = pd.read_csv("../data/stations_daily.csv")
data_hourly = pd.read_csv("../data/stations_hourly.csv")
```

### แสดงสถิติคร่าวๆ ของ ไฟล์รายวัน

```python
describe_col = data_daily.columns
describe_col = describe_col[2:] # ไม่เอา column แรก (station_id)
data_daily.describe()[describe_col]
```

### แสดงสถิติคร่าวๆ ของ ไฟล์รายชั่วโมง

```python
describe_col = data_hourly.columns
describe_col = describe_col[2:] # ไม่เอา column แรก (station_id)
data_hourly.describe()[describe_col]
```

### แสดงการกระจายข้อมูล

```python
import numpy as np
```

```python
describe_col
np.var(data_daily[describe_col], axis=0)
```

สังเกตว่าข้อมูลมีความแปรปรวนสูงมาก (ช่วงของค่ากว้างมากๆ)

```python
describe_col
np.var(data_hourly[describe_col], axis=0)
```

ทางไฟล์รายชั่วโมงก็เช่นกัน

```python
data_daily.isna().sum()
```

```python
data_hourly.isna().sum()
```
